```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("M16-Amazon-Challenge").getOrCreate()
```

```python
from pyspark import SparkFiles
url = "https://s3.amazonaws.com/amazon-reviews-pds/tsv/amazon_reviews_us_Sports_v1_00.tsv.gz"
spark.sparkContext.addFile(url)
df = spark.read.option("encoding", "UTF-8").csv(SparkFiles.get("amazon_reviews_us_Sports_v1_00.tsv.gz"), sep="\t", header=True, inferSchema=True)
df.show()
```

```python
clean_df = df.dropna()

clean_df.show()
```

```python
# Create the vine_table. DataFrame

vine_df = clean_df.select(["review_id", "star_rating", "helpful_votes", "total_votes", "vine", "verified_purchase"])

vine_df.show()
```

```python
# filter data and create new dataframe where total_votes count is equal to or greater than 20

from pyspark.sql.functions import col
total_votes_df = vine_df.filter(col("total_votes") >= 20)
total_votes_df.show()
```

```python
# filter and create new dataframe to retrieve all rows where the number of "helpful_votes" divided by "total votes" is => 50%

percent_vote_df = total_votes_df.withColumn('percent_vote', col('helpful_votes')/col('total_votes')).alias('percent_vote').filter(col("percent_vote")>= ".50")

percent_vote_df.show()
```

```python
# filter dataframe to retrieve rows where a review was written as part of vine program (paid)

paid_df = percent_vote_df.filter(col("vine") == "Y")

paid_df.show()
```

```python
# filter dataframe to retrieve rows where a review was not part of vine program (unpaid)

unpaid_df = percent_vote_df.filter(col("vine") == "N")

unpaid_df.show()
```

```python
# total number of reviews

total_reviews = vine_df.count()

total_reviews
```

```python
# number of 5-star reviews

five_star_reviews_df = vine_df.filter("star_rating == 5")
total_five_star_reviews = five_star_reviews_df.count()
total_five_star_reviews
```
